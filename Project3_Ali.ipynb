```python
# Import SQL Alchemy
from sqlalchemy import create_engine
import sqlite3
import pandas as pd
from sqlalchemy import Column, Integer, String, Float
from pathlib import Path
```

```python
database_path = "../Project-3-Group-3/fitness.sqlite"
Path(database_path).touch()

conn = sqlite3.connect(database_path)
csv_fitness = pd.read_csv("../Project-3-Group-3/Resources/exercise_dataset.csv")
csv_fitness.to_sql("Fitness", conn, if_exists='append', index=False)

conn.close()
```

```python
# Establish Connection
engine = create_engine("sqlite:///fitness.sqlite")
conn = engine.connect()
```

```python
try:
    # Execute a SQL query to fetch the data
    query = 'SELECT * FROM "Fitness"'  # Use quoted identifiers around the table name
    data = pd.read_sql_query(query, conn)
    print("Data successfully fetched from the database.")
except sqlite3.OperationalError as e:
    print("Error executing SQL query:", e)
finally:
    # Close the database connection
    conn.close()
```
